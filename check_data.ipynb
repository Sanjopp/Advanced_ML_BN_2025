```python
import numpy as np
import pickle
```

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        data_dict = pickle.load(fo, encoding='bytes')
    return data_dict
```

```python
data_dict = unpickle('DATA/cifar-10-batches-py/data_batch_1')
```

## Lecture simple des 5 batches CIFAR-10
Charge uniquement les 5 lots d'entraînement et affiche les shapes des données et labels.

```python
# Charge seulement les 5 lots d'entraînement et affiche les shapes
batches = [unpickle(f"DATA/cifar-10-batches-py/data_batch_{i}") for i in range(1, 6)]

data_all = np.vstack([b[b'data'] for b in batches])
labels_all = np.concatenate([b[b'labels'] for b in batches])

print("Data shape:", data_all.shape)
print("Labels shape:", labels_all.shape)
```

## Lecture du  test

```python
# Charge le lot de test et affiche les shapes
test_batch = unpickle('DATA/cifar-10-batches-py/test_batch')
test_data = np.array(test_batch[b'data'])
test_labels = np.array(test_batch[b'labels'])

print("Test data shape:", test_data.shape)
print("Test labels shape:", test_labels.shape)
```
