# Imports & setup

```python
import sys
from pathlib import Path
sys.path.append(str(Path("..").resolve()))

import torch
import matplotlib.pyplot as plt
from tqdm import trange

from models.mlp import MLP
from data_fashion import get_fashion_mnist_loaders
from train_utils import train_steps, evaluate
```

# Device & hyperparams

```python
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

EPOCHS = 10
LR = 0.1
BATCH_SIZE = 128
```

# Load Fashion-MNIST (auto-download)

```python
train_loader, test_loader = get_fashion_mnist_loaders(
    batch_size=BATCH_SIZE
)

x, y = next(iter(train_loader))
print(x.shape, y.shape)
```

# Lancement des expériences

```python
results = {}

for use_bn in [False, True]:
    name = "BN" if use_bn else "NoBN"
    model = MLP(input_dim=28 * 28, hidden_dims=(512, 256, 128), use_bn=use_bn).to(DEVICE)
    optimizer = torch.optim.SGD(model.parameters(), lr=LR)

    results[name] = {
        "loss": [],
        "grad_norm": [],
        "train_acc": [],
        "test_acc": []
    }

    for epoch in trange(EPOCHS, desc=f"Epochs ({name})"):
        train_steps(model, train_loader, optimizer, DEVICE, results[name])
        test_acc = evaluate(model, test_loader, DEVICE)
        results[name]["test_acc"].append(test_acc)
```

# Plot loss

```python
steps = range(1, len(results["NoBN"]["loss"]) + 1)

plt.plot(steps, results["NoBN"]["loss"], label="No BN", alpha=0.7)
plt.plot(steps, results["BN"]["loss"], label="With BN", alpha=0.7)
plt.xlabel("Step")
plt.ylabel("Training Loss")
plt.title("Training Loss per Step")
plt.legend()
plt.grid()
plt.show()
```

# Plot  accuracy

```python
# Calculer le nombre de steps par epoch pour placer les test_acc
steps = range(1, len(results["NoBN"]["train_acc"]) + 1)
steps_per_epoch = len(results["NoBN"]["train_acc"]) // EPOCHS
test_steps = [(i + 1) * steps_per_epoch for i in range(EPOCHS)]

plt.plot(steps, results["NoBN"]["train_acc"], label="Train No BN", alpha=0.6)
plt.plot(steps, results["BN"]["train_acc"], label="Train BN", alpha=0.6)
plt.plot(test_steps, results["NoBN"]["test_acc"], marker='o', label="Test No BN", linewidth=2)
plt.plot(test_steps, results["BN"]["test_acc"], marker='o', label="Test BN", linewidth=2)
plt.xlabel("Step")
plt.ylabel("Accuracy")
plt.title("Train & Test Accuracy per Step")
plt.legend()
plt.grid()
plt.show()
```

# Plot gradient norms

```python
steps = range(1, len(results["NoBN"]["grad_norm"]) + 1)

plt.plot(steps, results["NoBN"]["grad_norm"], label="No BN", alpha=0.7)
plt.plot(steps, results["BN"]["grad_norm"], label="With BN", alpha=0.7)
plt.xlabel("Step")
plt.ylabel("Gradient Norm")
plt.title("Gradient Norm Evolution per Step")
plt.legend()
plt.grid()
plt.show()
```
